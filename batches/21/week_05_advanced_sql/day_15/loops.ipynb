```sql
SELECT prod_id,stock_qty
FROM products.inventory
WHERE
    stock_qty < 10;
```

```sql

Do $$
DECLARE
    i record;
BEGIN
FOR i IN
    SELECT prod_id,stock_qty
    FROM products.inventory
    WHERE
        stock_qty < 10
loop
    raise notice 'Low Stock Product : % has only % left !!',i.prod_id,i.stock_qty;
    RAISE notice E'----------\n';
END LOOP;
end;
$$ LANGUAGE plpgsql;
```
